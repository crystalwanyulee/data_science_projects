```python
color_dict = {'g':'#35b494', 'y': '#f7b025', 'r':'#ea5345', 'b':'#79b9ce', 'title':'#595959'}
clusters_palette = ['#f7b025', 'lightcoral', '#35b494', '#79b9ce', 'silver']
text_palette = ['#F7A025', color_dict['r'], 'seagreen', 'steelblue', 'grey']
#clusters_palette = ['coral','#f4b1b2','#f7b025', '#35b494', '#79b9ce', 'silver']
sns.palplot(clusters_palette)
sns.palplot(text_palette)
```

```python
def plot_style(fontsize=14, style='bmh', facecolor='white', textcolor='#595959'):
    plt.style.use(style)
    mpl.rcParams['axes.facecolor'] = facecolor
    plt.rc('font', **{'size'   : fontsize})
    plt.rcParams.update({i:textcolor for i in ["ytick.color", "xtick.color","axes.labelcolor","text.color"]})
```

```python
from itertools import compress

def hide_spines(ax, top=True, bottom=True, left=True, right=True):
    sides = ['top', 'bottom', 'left', 'right']
    mask = [top, bottom, left, right]
    for i in list(compress(sides, mask)):
        ax.spines[i].set_visible(False)
```

# Data

```python
df = pd.read_csv('attribution data.csv') 
# Correct the datatype of the time variable
df['time'] = pd.to_datetime(df['time'])
df.head()
```

# EDA

```python
df_temp = pd.pivot_table(df, index='cookie', columns=['channel'], values='time', aggfunc=pd.Series.nunique)
```

```python
df_temp = df.groupby('channel')['time'].count().reset_index().rename({'time':'count'}, axis=1).sort_values('count', ascending=False).set_index('channel')
df_temp['percentage'] = df_temp['count']/df_temp['count'].sum()

plot_style(fontsize=16)
fig, ax1 = plt.subplots(figsize=[10,6], dpi=80)
df_temp['count'].plot.bar(ax=ax1, color=color_dict['b'], zorder=10)

ax1.set_yticks(np.arange(0,20e04+1,5e04))
ax1.set_yticklabels([str(i)+'K' for i in np.arange(0,201,50)])
ax1.set_xticklabels([i.replace(' ', '\n') for i in df_temp.index])

ax1.set_xlabel('')
ax1.set_ylabel('Impression', labelpad=10)
ax1.set_title('Impression by Channels', pad=20)

ax1.tick_params(rotation=0, pad=10, length=0)
ax1.grid(False)
ax1.grid(axis='y', which='major', color='silver', zorder=5)
hide_spines(ax1, bottom=False)

for index, row in df_temp.reset_index().iterrows():
    string='{0:.1%}'.format(row['percentage'])
    ax1.text(index, row['count']*0.5, string, ha='center', va='center', color='white', zorder=15)

plt.show()
#plt.savefig('Impression by Channels', facecolor='white', bbox_inches="tight", dpi=85)
```

```python
df_temp = df['conversion'].value_counts()

plot_style(fontsize=16)
fig, ax1 = plt.subplots(figsize=[10,6], dpi=80)

df_temp.plot.bar(ax=ax1, zorder=10, color=[color_dict['b'], color_dict['r']])
ax1.set_xticklabels(['Non-Conversion', 'Conversion'])

#ax1.set_xlim(-0.5,1.5)
#ax1.set_ylim(0,6e04)
ax1.set_yticklabels('')
ax1.tick_params(rotation=0, pad=10, length=0)
ax1.set_xlabel('')
ax1.set_title('The Number of Conversions', pad=20)

hide_spines(ax1, bottom=False)

colors = ['white', color_dict['r']]
for index, row in df_temp.reset_index().iterrows():
    string = '{0:,}\n({1:.0%})'.format(row['conversion'], row['conversion']/df_temp.sum())
    ax1.text(index, 8e04, string, ha='center', va='center', color=colors[index], zorder=15, fontsize=20)

ax1.grid(False)
plt.show()
#plt.savefig('The Number of Conversions', facecolor='white', bbox_inches="tight", dpi=85)
```

# Interpret Modeling Results

```python
transition = pd.read_csv('data/markov_transition.csv')
markov = pd.read_csv('data/markov_conversion.csv')
removal = pd.read_csv('data/markov_removal.csv')
heuristic = pd.read_csv('data/heuristic_model.csv')
markov_high = pd.read_csv('data/markov_conversion_high_order.csv')
```

```python
markov['conversion_ratio'] = markov['total_conversions']/markov['total_conversions'].sum()
markov = markov.sort_values('total_conversions', ascending=False).reset_index()


plot_style()
fig, ax1 = plt.subplots(figsize=[10,6], dpi=80)
ax2 = ax1.twinx()
markov.set_index('channel_name')['total_conversions'].plot.bar(ax=ax1, color='#ef846d',zorder=10)
#markov.set_index('channel_name')['total_conversion_value'].plot(ax=ax2, color='#ef846d', linewidth=3, zorder=15)
ax1.set_ylim(0,6000)
ax2.set_ylim(0,40000)
ax2.set_yticks(np.arange(0,40001, 10000))
ax1.set_yticks(np.arange(0,6001, 1500))
ax1.set_xticklabels([i.replace(' ', '\n') for i in markov.channel_name])
ax1.set_xlabel('')
ax2.set_yticklabels('')

for ax in [ax1, ax2]:
    ax.grid(False)
    ax.tick_params(axis='both', length=0, which='major', rotation=0, pad=10)
    hide_spines(ax, bottom=False)

ax1.grid(axis='y', which='major', color='silver', zorder=5)

for index, row in markov.iterrows():
    string='{0:.1%}'.format(row['conversion_ratio'])
    ax1.text(index, row['total_conversions']*0.5, string, ha='center', va='center', color='white', zorder=15)
    
plt.title('Conversions (Markov Chain)', fontsize=20, color=color_dict['title'], y=1.05)

plt.show()
#plt.savefig('Convertions Computed by Markov Chain', facecolor='white', bbox_inches="tight", dpi=85)
```

```python
compare_conv = pd.merge(heuristic.loc[:, ['channel_name', 'first_touch_conversions', 'last_touch_conversions', 'linear_touch_conversions']], 
                        markov.loc[:, ['channel_name', 'total_conversions']].rename({'total_conversions':'markov_chain_conversions'}, axis=1), on='channel_name')
compare_conv = pd.melt(compare_conv, id_vars='channel_name')
compare_conv = pd.pivot(compare_conv, index='variable', columns='channel_name', values='value').reset_index()
compare_conv['variable'] = compare_conv['variable'].apply(lambda x: ' '.join([i.capitalize() for i in x.split('_')[:2]]))
compare_conv = compare_conv.iloc[:, [0,1,5,4,2,3]].reset_index(drop=True)
compare_conv
```

```python
import matplotlib.lines as mlines
colors = [clusters_palette[i] for i in [0,3,2,1]]
text_colors = ['#F7A025', 'steelblue', 'seagreen', color_dict['r']]
plot_style(fontsize=16)
fig, ax = plt.subplots(figsize=[8,6], dpi=90)
parallel_coordinates(compare_conv, 'variable', color=colors, ax=ax, axvlines=False, linewidth=3)
hide_spines(ax, right=False, left=False)

for i in range(1,5):
    ax.axvline(x=i, linewidth=0.5, color=color_dict['title'])

for side in ['right', 'left']:
    ax.spines[side].set_color(color_dict['title'])
    ax.spines[side].set_linewidth(0.5)
    
ax.tick_params(axis='x', which='both', length=0, pad=10)
ax.tick_params(axis='y', which='both', length=0, pad=10, direction='out', color='grey')
ax.legend().remove()
ax.set_xticklabels([col.replace(' ', '\n') for col in compare_conv.columns[1:]])
ax.set_yticks(np.arange(1000, 6000, 1000))

ax.set_ylim(0,6001, 1000)
ax.grid(True, color='silver')
   
legend_labels = []
for i in range(4):
    legend_labels.append(mlines.Line2D([], [], color=colors[i], marker='.', 
                         linestyle='None', markersize=13, label=compare_conv.variable[i]))

plt.legend(handles = legend_labels, bbox_to_anchor=(1, .7), fontsize=16, edgecolor='white', handletextpad=0.05)  
    
plt.title('Conversions by Channels', fontsize=20, color=color_dict['title'], y=1.05)

plt.show()
#plt.savefig('The Number of Convertions by Channels', facecolor='white', bbox_inches="tight", dpi=80)
```
