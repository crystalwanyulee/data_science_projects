```python
retail[['Quantity','UnitPrice']].describe()
```

* From the above table, it is noteworthy that Quantity and Unit Price have negative values and extreme values. Negative values is due to cancellation of transactions and thus it will not affect the calculation of sales. However, extreme values might cause influences on the results of clustering, and thus I will deal with them after finishing RFM analysis.

```python
# Remove rows with unknown customer ID
retail = retail.loc[retail['CustomerID'].notna()]

# Remove duplicate rows
retail[~retail.duplicated()]

# Change the datatype of CustomerID
retail['CustomerID'] = retail['CustomerID'].astype('int').astype('str') 
```

```python
# Find cancelled orders and thier original orders
cancelled = retail.loc[retail['InvoiceNo'].str.startswith('C', na=False), 'InvoiceNo'].unique()
original =  [i.replace('C', '') for i in cancelled]

# Filter out the cancelled orders
retail = retail[~retail['InvoiceNo'].isin(cancelled)]
                                         
# Reduce revenue in the original orders to 0
retail.loc[retail['InvoiceNo'].isin(original), 'UnitPrice'] = 0                                         
```

```python
# Calculating the total amount of payment for each InvoiceNo
retail['revenue'] = retail['Quantity']*retail['UnitPrice']
```

```python
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
```

# RFM Analysis

```python
# Compute RFM
today = pd.to_datetime('2011-12-10', format='%Y-%m-%d')
rfm = retail.groupby('CustomerID').agg({'InvoiceDate': {'recency': lambda x: (np.max(x)- today).days},
                                        'InvoiceNo': {'frequency': pd.Series.nunique},
                                        'revenue': {'monetary': np.sum}})
rfm.columns = rfm.columns.droplevel()
rfm.head()
```

```python
rfm.head().style.set_table_attributes('style="font-size: 20px"')
```

```python
fig, ax = plt.subplots(ncols=3, figsize=[12,4], dpi=100, sharey=True)
for i in range(3):
    sns.distplot(rfm.iloc[:, i], ax=ax[i], color=color_dict['r'], kde=False, norm_hist=False)
    ax[i].tick_params(axis='y', which='both', length=0, pad=10)
    ax[i].set_title(rfm.columns[i].capitalize(), fontsize=16, y=1.03, color=color_dict['title'])
    ax[i].set_xlabel('')
    ax[i].set_yticks(np.arange(0,3500,1000))
    ax[i].grid(color='lightgrey', which='major', axis='y', linewidth=0.3)
    
plt.tight_layout()    
plt.savefig('Distribution of RFM before removing outliers', facecolor='white', bbox_inches="tight", dpi=150)
```

```python
outlier_filter = (np.abs((rfm - rfm.mean()) / rfm.std()) < 3).sum(axis=1) == 3
rfm_clean = rfm.iloc[outlier_filter.values]
print('removed '+ str(len(rfm)-len(rfm_clean)) +' outliers')
rfm_clean.head()
```

```python
plt.style.use('ggplot')
mpl.rcParams['axes.facecolor'] = 'white'
fig, ax = plt.subplots(ncols=3, figsize=[12,4], dpi=100, sharey=True)
for i in range(3):
    sns.distplot(rfm_clean.iloc[:, i], ax=ax[i], color=color_dict['r'], kde=False, norm_hist=False, bins=20)
    ax[i].tick_params(axis='y', which='both', length=0, pad=10)
    ax[i].set_title(rfm.columns[i].capitalize(), fontsize=16, y=1.03, color=color_dict['title'])
    ax[i].set_xlabel('')
    ax[i].set_yticks(np.arange(0,3500,1000))
    ax[i].grid(color='lightgrey', which='major', axis='y', linewidth=0.3)
    
plt.tight_layout()    
plt.savefig('Distribution of RFM after removing outliers', facecolor='white', bbox_inches="tight", dpi=150)
```

# Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
rfm_scaled = StandardScaler().fit_transform(rfm_clean.iloc[: , :3])
```

```python
plt.style.use('bmh')
def find_best_k(x, kmax=10):
    wss = []
    sil = []
    for k in range(2, kmax+1):
        km = KMeans(n_clusters = k,
                        n_init=10, max_iter=300,
                        tol=1e-04, random_state=0)
        km.fit(x)
        sil.append(silhouette_score(x, km.labels_, metric = 'euclidean'))
        wss.append(km.inertia_)
    fig, ax = plt.subplots(nrows=2, figsize=[10, 8])
    sns.lineplot(x=np.arange(2,11), y=wss, ax=ax[0], color='#35b494')
    ax[0].set_xlabel('Number of clusters')
    ax[0].set_ylabel('WSS')
    ax[0].set_title('Finding the Best Number of Clusters', fontsize=16, y=1.05, color='#595959')
    sns.lineplot(x=np.arange(2,11), y=sil, ax=ax[1], color='#41b0e5')
    ax[1].set_xlabel('Number of clusters')
    ax[1].set_ylabel('Silhouette Score') 
#    ax[1].grid(color='lightgrey', which='major', axis='both', linewidth=0.3)
```

```python
mpl.rcParams['axes.facecolor'] = 'white'
find_best_k(rfm_scaled)
plt.savefig('The best k for clustering', facecolor='white', bbox_inches="tight", dpi=150)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
rfm_pc = pca.fit_transform(rfm_scaled)
rfm_clean['principal component 1'], rfm_clean['principal component 2'] = rfm_pc[:,0], rfm_pc[:,1]
    
```

```python
fig, ax = plt.subplots(ncols=4, figsize=[18,4])
plt.subplots_adjust(wspace=0.5)
# Choose 4 as the optimal value of k

for index, n in enumerate(list(range(2,6))):
    km = KMeans(n_clusters = n,
            n_init=10, max_iter=300,
            tol=1e-04, random_state=0)
    km.fit(rfm_scaled)
    rfm_clean['km_cluster'] = km.labels_
    sns.scatterplot(x='principal component 1', y='principal component 2', hue='km_cluster', data=rfm_clean, palette='Set2', ax=ax[index])
    ax[index].set_title('k={:}'.format(index+2))
    ax[index].legend().remove()
```

```python
# Define a function to calculate how many days there are between two dates
from datetime import date
def date_diff(d1, d2):
    d1 = date.strftime(d1, "%Y-%m-%d")
    d2 = date.strftime(d2, "%Y-%m-%d")
    
    return d1-d2
```
