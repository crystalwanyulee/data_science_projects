```python
color_dict = {'g':'#35b494', 'y': '#f7b025', 'r':'#ea5345', 'b':'#79b9ce', 'title':'#595959'}
clusters_palette = ['#f7b025', 'lightcoral', '#35b494', '#79b9ce', 'silver']
text_palette = ['#F7A025', color_dict['r'], 'seagreen', 'steelblue', 'grey']
#clusters_palette = ['coral','#f4b1b2','#f7b025', '#35b494', '#79b9ce', 'silver']
sns.palplot(clusters_palette)
sns.palplot(text_palette)
```

```python
def plot_style(fontsize=14, style='bmh'):
    plt.style.use(style)
    mpl.rcParams['axes.facecolor'] = 'white'
    plt.rc('font', **{'size'   : fontsize})
    plt.rcParams.update({i:color_dict['title'] for i in ["ytick.color", "xtick.color","axes.labelcolor","text.color"]})
```

```python
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value
        patch.set_width(new_value)
        patch.set_x(patch.get_x() + diff * .5)
```

```python
from itertools import compress

def hide_spines(ax, top=True, bottom=True, left=True, right=True):
    sides = ['top', 'bottom', 'left', 'right']
    mask = [top, bottom, left, right]
    for i in list(compress(sides, mask)):
        ax.spines[i].set_visible(False)
```

```python
def newline(p1, p2, color='black'):
    ax = plt.gca()
    l = mlines.Line2D([p1[0],p2[0]], [p1[1],p2[1]], color=color, zorder=0.5, linewidth=1.5)
    ax.add_line(l)
    return l
```

# Part 1. Feature Engineering by RFM Analysis

* Read the dataset I have cleaned in [the previous article](https://crystallee7816.github.io/Website/online-retail-customer-segmentation.html)

```python
retail = pd.read_csv('retail_clean.csv')
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'])
retail.head()
```

### 1-1. Calculate RFM

* To calculate recency, I defined a snapshot date as December 9, 2011, which is the next day of the last order.

* Since only recency prefers lower values, here I took negative recency to make all features have the same evaluation.

```python
# Specify the snapshot date
today = pd.to_datetime('2011-12-10', format='%Y-%m-%d')
# Compute RFM
rfm = retail.groupby('CustomerID').agg(lifetime=('InvoiceDate', lambda x: (x.max() - x.min()).days),
                                       recency=('InvoiceDate', lambda x: (x.max() - today).days),
                                       frequency=('InvoiceNo', pd.Series.nunique),
                                       monetary=('revenue', 'sum'))

rfm.head()
```

The new structure exhibits each customer's purchase behavior in each row. For instance, Customer ID 12347 had been alive for 1 year and his or her last purchase was 3 days ago. Plus, he or she had created 7 orders and brought about 4,310 dollars. 

### 1-2. Checking Distributions of New Features

```python
# Check distributions of features
plot_style()

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=[10,8], dpi=75, sharey=True)
col_idx = 0

for i in range(2):
    for j in range(2):
        sns.distplot(rfm.iloc[:, col_idx], ax=ax[i,j], color=color_dict['r'], kde=False, norm_hist=False)
        ax[i,j].tick_params(axis='y', which='both', length=0)
        ax[i,j].set_xlabel('123', color='white')
        ax[i,j].set_title(rfm.columns[col_idx].capitalize(), fontsize=20, color=color_dict['title'], y=1.1)
        ax[i,j].set_yticks(np.arange(0,5000,1000))
        col_idx += 1

ax[1,1].set_xticklabels([str(i)+'K' for i in np.arange(-50,251,50)])

plt.tight_layout()    
plt.savefig('Distribution of RFM before removing outliers', facecolor='white', bbox_inches="tight", dpi=75)
```

* Lifetime is a little bit left-skewed, meaning that only a few customers were alive for a long time.

* Recency is slightly right-skewed. Generally, it is good news for e-commerce because most customers are still active.

* For frequency and monetary, values are centered on the left side of the plot. That is, some extreme values lead to skewness. Since they may influence the results of clustering, it is better to remove those outliers.

### 1-3. Remove Outliers

* Definition of outliers: the observations with the value of frequency or monetary exceeding the value in the 99.5 percentile.

* In this case, lifetime and recency are not appropriate to remove outliers, so I did not consider them in recognition of outliers.

```python
col_selected = ['frequency','monetary']
outlier_filter = (rfm[col_selected] > rfm[col_selected].quantile(0.995)).sum(axis=1) == 0 
rfm_clean = rfm.loc[outlier_filter.values]
print('removed '+ str(len(rfm)-len(rfm_clean)) +' outliers')
```

### 1-4. Examing Distributions Again

```python
# Check distributions of features again
plot_style()

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=[10,8], dpi=75, sharey=True)
col_idx = 0

for i in range(2):
    for j in range(2):
        sns.distplot(rfm_clean.iloc[:, col_idx], ax=ax[i,j], color=color_dict['r'], kde=False, norm_hist=False)
        ax[i,j].tick_params(axis='y', which='both', length=0)
        ax[i,j].set_xlabel('123', color='white')
        ax[i,j].set_title(rfm_clean.columns[col_idx].capitalize(), fontsize=20, color=color_dict['title'], y=1.1)
        ax[i,j].set_yticks(np.arange(0,5000,1000))
        col_idx += 1
    
ax[1,1].set_xticklabels([str(i)+'K' for i in np.arange(-50,251,50)])

plt.tight_layout()    
plt.savefig('Distribution of RFM after removing outliers', facecolor='white', bbox_inches="tight", dpi=75)
```

* Although skewness still exists, the distributions look better now, and features are able to be applied in the subsequent clustering task.

# Part 2. Segmentation by K-Means Clustering

### 2-1. Loading packages and Normalizing Data

Since K-means is a distance-based model, the vast difference between scales of variables leads to biased results. Thus, we have to normalize data before deploying the algorithm. 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_clean.iloc[: , :4])
```

### 2-2. Choosing The Best Number of K

K-means requires us to define the number of clusters in advance. To find the optimal one, I tried different values from 2 to 10 to run the model and evaluated the results through two metrics, WSS and Silhouette score. 

```python
def find_best_k(x, kmax=10):
    wss = []
    sil = []
    for k in range(2, kmax+1):
        km = KMeans(n_clusters = k,
                        n_init=10, max_iter=300,
                        tol=1e-04, random_state=0)
        km.fit(x)
        sil.append(silhouette_score(x, km.labels_, metric = 'euclidean'))
        wss.append(km.inertia_)
    fig, ax = plt.subplots(nrows=2, figsize=[10, 8])
    sns.lineplot(x=np.arange(2,11), y=wss, ax=ax[0], color='#35b494')
    ax[0].set_xlabel('')
    ax[0].set_ylabel('WSS')
    ax[0].set_title('Finding the Best Number of Clusters', fontsize=20, y=1.1, color='#595959')
    sns.lineplot(x=np.arange(2,11), y=sil, ax=ax[1], color='#41b0e5')
    ax[1].set_xlabel('Number of clusters')
    ax[1].set_ylabel('Silhouette Score') 
```

```python
plot_style()
find_best_k(rfm_scaled)
plt.savefig('The best k for clustering', facecolor='white', bbox_inches="tight", dpi=75)
```

WSS measures how compressed each cluster is, and thus it is better to have a low WSS. Yet, the lower WSS is, the more clusters the model produces. Typically, it is unnecessary to have lots of small groups because it is hard to sketch some patterns. Thus, it is common to use the elbow method to find a turning point where WSS starts to decline slowly. In our case, the turning point is probably located at 4 or 5.

Another indicator is the Silhouette score, which measures the distance between clusters and tells us how well the separation is. Generally, it is better to choose the number with the highest Silhouette score. From this point, **the best number of clusters is 5**.



### 2-3. Evaluation of the Clustering Results
Now, I will use k=5 to apply the K-Means algorithm again. Since the algorithm could not identify which cluster is better, I compared the average values in each variable and ranked each cluster as the following.

```python
# Use the optimal k=5 to implement KMeas model
km = KMeans(n_clusters = 5,
            n_init=10, max_iter=300,
            tol=1e-04, random_state=0)
km.fit(rfm_scaled)

# Assign the clustering results back into the dataframe
rfm_clean['km_cluster'] = km.labels_

# Identify differences between clusters and rank clusters
rfm_cluster_mean = rfm_clean.groupby('km_cluster').mean().reset_index()
rfm_cluster_mean['rank'] = rfm_cluster_mean.iloc[:,1:].rank().sum(axis=1).rank(ascending=False).astype('category')
rfm_cluster_mean['count'] = rfm_clean.groupby('km_cluster')['lifetime'].count()
rfm_cluster_mean = rfm_cluster_mean.sort_values('rank').round().reset_index(drop=True)

# Relabel clusters
new_labels = ['VIP', 'Loyal', 'Casual', 'Novice', 'Probably Dead']
rfm_cluster_mean['label'] = new_labels

# Print the results
rfm_cluster_mean.set_index('label').drop(['km_cluster','rank'], axis=1)
```

# Part 3. Interpretation of Clusters

### 3-1. Separation
To inspect the relationships between cluster, I map all data points into 2 dimesntional space through princial component analysis and depict their positions in a scatter plot. 

```python
# Conduct PCA for plotting
from sklearn.decomposition import PCA
rfm_pc = PCA(n_components=2).fit_transform(rfm_scaled)
rfm_clean['pc_1'] = rfm_pc[:,0]
rfm_clean['pc_2'] = rfm_pc[:,1]

# Assign the ranks back into the dataframe
rfm_clean['rank'] = 999
for i in range(5):
    rfm_clean.loc[rfm_clean['km_cluster'] == i, 'rank'] = rfm_cluster_mean.loc[rfm_cluster_mean['km_cluster'] == i, 'rank'].values
```

```python
plot_style(fontsize=12)
fig, ax = plt.subplots(figsize=[7,6], dpi=100)

sns.scatterplot(x='pc_1', y='pc_2', hue='rank', data=rfm_clean, palette=clusters_palette, ax=ax, s=60, alpha=0.8)

ax.set_title('Separation of Clusters', fontsize=20, color='#595959', y=1.05)
ax.set_xlabel('PC 1', labelpad=10)
ax.set_ylabel('PC 2', labelpad=10)

ax.tick_params(axis='y', which='both', length=0,  pad=10)
hide_spines(ax)

legend_labels = []
for i in range(5):
    legend_labels.append(mlines.Line2D([], [], color=clusters_palette[i], marker='.', 
                                      linestyle='None', markersize=13, label=new_labels[i]))

plt.legend(handles = legend_labels, bbox_to_anchor=(1, .7), fontsize=16, edgecolor='white', handletextpad=0.05)    

plt.savefig('Separation when k=5', facecolor='white', bbox_inches="tight", dpi=85)
```

We can recognize different cluster from at least one feature, and clusters can be roughly divided into three levels:

* **Top VIP and Regular Customer:** All features are above the mean. They are alive for a long time and are active all the time. Their contributions are high enough to be loyal customers. The primary differences between subgroups are frequency and monetary.

* **Medium Causal Buyer:** All indicators are in the middle. Although they are alive for a long time and are active all the time, they are not loyal to the business. Thus, they visit once in a while and are not willing to spend a lot each time. 

* **Bottom Novice and Probably Dead:** Almost all values are below the mean. Two groups behave similarly, but they only differ in recency. Roughly speaking, Probably dead buyers are not active now, but novice customers may still alive.
